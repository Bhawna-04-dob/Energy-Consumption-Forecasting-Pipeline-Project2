```python
%python


from pyspark.sql.utils import AnalysisException

spark.sql("USE CATALOG Silver_Catalog")
spark.sql("USE SCHEMA Silver_SCH")

tables = {

    "Silver_Energy_Usage": "household_id",

    "Silver_Weather": "household_id",

    "Silver_Device_Metrics": "household_id",

    "Silver_Grid_Load": "household_id",

    "Silver_Tariff_Metrics": "household_id"

}

for table, zorder_col in tables.items():

    try:

        print(f"\n==============================")
        print(f"Processing : {table}")
        print(f"==============================")

        # OPTIMIZE
        spark.sql(f"""
        OPTIMIZE Silver_Catalog.Silver_SCH.{table}
        """)

        print(f"✅ {table} OPTIMIZE Completed")

        # ZORDER
        spark.sql(f"""
        OPTIMIZE Silver_Catalog.Silver_SCH.{table}
        ZORDER BY ({zorder_col})
        """)

        print(f"✅ {table} ZORDER Completed")

        # VACUUM
        spark.sql(f"""
        VACUUM Silver_Catalog.Silver_SCH.{table}
        RETAIN 168 HOURS
        """)

        print(f"✅ {table} VACUUM Completed")

        # HISTORY
        print(f"History for {table}")

        spark.sql(f"""
        DESCRIBE HISTORY Silver_Catalog.Silver_SCH.{table}
        """).show(truncate=False)

    except Exception as e:

        print(f"❌ Error while optimizing {table}")

        print(str(e))

print("===================================")
print("Silver Optimization Completed")
print("===================================")
```
